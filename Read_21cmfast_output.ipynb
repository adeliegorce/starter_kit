## 21CMFAST cube

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

21CMFAST outputs are saved as h5df files. Here is how to read them:

```python
import h5py 
filename = 'mycube.h5'
f = h5py.File(filename, "r")
```

```python
# ionisation field
xion = 1. - np.array(f['IonizedBox']['xH_box'])
N = xion.shape[0]

plt.figure()
plt.imshow(xion[:,:,N//2])
plt.colorbar()
```

```python
# brightness temperature
cube21 = np.array(f['BrightnessTemp']['brightness_temp']) #mK
cube21 -= np.mean(cube21)

plt.figure()
plt.imshow(cube21[:,:,N//2])
plt.colorbar()
```
